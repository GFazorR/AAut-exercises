```python
import pandas as pd
import numpy as np

clusters_3_df = pd.read_csv("../Datasets/3-clusters.csv", header=1)
# display(clusters_3_df)
dbscan_df = pd.read_csv("../Datasets/dataset-DBSCAN.csv", header=1)
# display(dbscan_df)
cure_df = pd.read_csv("../Datasets/CURE-complete.csv", header=1)
# display(cure_df)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
fig.subplots_adjust(top=1)
ax.set_title("Dataset n.1 of data points")
ax.set_xlabel(clusters_3_df.columns[0])
ax.set_ylabel(clusters_3_df.columns[1])

plt.plot(clusters_3_df["x"], clusters_3_df["y"], ". ", markersize=1)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
fig.subplots_adjust(top=1)
ax.set_title("Dataset n.1 of data points")
ax.set_xlabel(cure_df.columns[0])
ax.set_ylabel(cure_df.columns[1])

plt.plot(cure_df["x"], cure_df["y"], ". ", markersize=1)

plt.show()
```

```python
from sklearn.cluster import KMeans

k = 3
kmc_1 = KMeans(n_clusters=k, random_state=0).fit(clusters_3_df.to_numpy())

for i in range(clusters_3_df.shape[0]):
    print(f"Example n.{i}=({clusters_3_df['x'][i]},{clusters_3_df['y'][i]})")
    print(f"in cluster n.{kmc_1.labels_[i]}")
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
fig.subplots_adjust(top=1)
ax.set_title("Dataset n.1 of data points")
ax.set_xlabel(clusters_3_df.columns[0])
ax.set_ylabel(clusters_3_df.columns[1])

color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

for clu in range (k):
    data_list_x = [clusters_3_df["x"][i] for i in range(clusters_3_df.shape[0]) if kmc_1.labels_[i] == clu]
    data_list_y = [clusters_3_df["y"][i] for i in range(clusters_3_df.shape[0]) if kmc_1.labels_[i] == clu]
    plt.scatter(data_list_x, data_list_y, s=2, edgecolors='none', c=color[clu], alpha=0.5)

plt.show()
```
